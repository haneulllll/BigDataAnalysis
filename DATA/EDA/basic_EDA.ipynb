```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn 스타일 설정
sns.set_theme(style="whitegrid")

# CSV 파일 로딩 
file_path = "Dynamic_20200310.csv"
use_cols = ["MMSI", "일시", "위도", "경도", "SOG", "COG", "Heading"]
dtype_dict = {
    "MMSI": "string", "위도": "float32", "경도": "float32",
    "SOG": "float32", "COG": "float32", "Heading": "float32"
}
df = pd.read_csv(file_path, skiprows=2,encoding="cp949", usecols=use_cols, dtype=dtype_dict, parse_dates=["일시"])

# 1. 데이터 개요 출력
print(df.info())
print(df.describe())
```

```python
# 2. SOG, COG, Heading 분포 시각화
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df["SOG"], bins=50, kde=True, ax=axes[0])
axes[0].set_title("SOG(knots)")
axes[0].set_xlabel("SOG (knots)")

sns.histplot(df["COG"], bins=50, kde=True, ax=axes[1])
axes[1].set_title("COG(degrees)")
axes[1].set_xlabel("COG (degrees)")

sns.histplot(df["Heading"], bins=50, kde=True, ax=axes[2])
axes[2].set_title("Heading(degrees)")
axes[2].set_xlabel("Heading (degrees)")

plt.show()
```

```python
# 3. 위도-경도 분포 시각화
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["longitude"], y=df["latitude"], alpha=0.3, s=10)
plt.title("latitude - longitude")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

```python
# 4. 시간대별 AIS 데이터 수 (히스토그램)
df["hour"] = df["일시"].dt.hour
plt.figure(figsize=(10, 5))
sns.histplot(df["hour"], bins=24, kde=False)
plt.title("AIS_DATA COUNT per TIME")
plt.xlabel("TIME (Hour)")
plt.ylabel("COUNT")
plt.xticks(range(0, 24))
plt.show()
```

```python
# 5. Pairplot: SOG, COG, Heading 간 관계
sns.pairplot(df[["SOG", "COG", "Heading"]])
plt.show()
```
